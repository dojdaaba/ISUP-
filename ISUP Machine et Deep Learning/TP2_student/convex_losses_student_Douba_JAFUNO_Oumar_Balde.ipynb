```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Plot some convex surrogates for the 0-1 loss

Here, we propose to plot the following losses compared to the 0-1 loss:
+ Hinge loss (SVM) : $\ell(y,y') = (1-yy')_+$
+ Quadratic Hinge loss :  $\ell(y,y') = \frac{1}{2}(1-yy')_+^2$
+ Huber loss : $ \ell(y,y') = -4 yy'\mathbb{1}_{yy'<-1}+  (1-yy')_+^2\mathbb{1}_{yy'\geq-1}$
+ Logit loss : $\ell (y , y') = \log (1+e^{-yy'} )$

```python
# Hinge
x = np.arange(-3, 3, step = 0.1)
y = x.copy()
for i in range( len(x) ):
    if x[i] < 1:
        y[i] = 1-x[i]
    if x[i] >= 1:
        y[i]=0

plt.plot(x,y);
```

```python
# Quadratic Hinge Loss
x = np.arange(-3, 3, step = 0.1)
y = x.copy()
for i in range( len(x) ):
    if x[i] < 1:
        y[i] = 0.5*(1-x[i])*(1-x[i])
    if x[i] > 1:
        y[i]=0

plt.plot(x,y);
```

```python
# Logit
x1 = np.arange(-3, 3, step = 0.1)
y1 = np.log(1 + np.exp(-x1))

plt.plot(x1, y1);

```

```python
#Huber loss
x = np.arange(-3, 3, step = 0.1)
y = x.copy()
for i in range( len(x) ):
    if x[i] < -1:
        y[i] = -4*x[i]
    if x[i] >=-1:
        y[i]=(1-x[i])*(1-x[i])

plt.plot(x,y);

```
